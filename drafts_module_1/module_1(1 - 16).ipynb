# Вопрос 13. У какого режиссёра самые большие суммарные кассовые сборы?


```python
def director_revenue(df):
    summ = Counter()
    for i in range(len(df)):
        for j in df.iloc[i].director.split('|'):
            summ[j] += df.iloc[i].revenue
    return pd.DataFrame.from_dict(summ, orient='index', columns = ['Summ'])
display(director_revenue(data[data.revenue > 0]).sort_values(by=['Summ'], ascending = False))
```

В этой функции создается словарь, в котором ключи - имя режиссера 
(если режиссеров несколько, то они разделяются). Вы проходитесь по датафрейму и 
заполняете словарь, увеличивая значение того или иного ключа в каждой итерации 
цикла вот этой строкой:
            summ[j] += df.iloc[i].revenue
После чего возвращаете датафрейм, который построен исходя из данных словаря, ключи 
словаря стали индексами датафрейма, а значения - значениями колонки Summ

# 15. Фильмы с каким актером принесли самые высокие кассовые сборы в 2012 году?

```python
grouped_2012 = data[(data['release_year'] == 2012)]
grouped_2012.head(2)
```

```python
def cast_revenue(df):
    summ = Counter()
    for i in range(len(df)):
        for j in df.iloc[i].cast.split('|'):
            summ[j] += df.iloc[i].revenue
    return pd.DataFrame.from_dict(summ, orient='index', columns = ['Summ'])
display(cast_revenue(grouped_2012[grouped_2012.revenue > 0]).sort_values(by=['Summ'], ascending = False))
```

```python
# Другой вариант
# берем копию исходного датафрейма:
df = data.copy()
```

```python
#Создаем в ней столбец с разделенными значениями актеров:
df['new_cast'] = data[data.release_year == 2012]['cast'].str.split('|')
df
```

```python
#Преобразуем каждый элемент списка в строку:
df = df.explode('new_cast')
df
```

```python
# группируем по значению выручки:
df.groupby(by = 'new_cast')['revenue'].agg('sum').sort_values(ascending=False)
```

```python
# Если использовать код из предыдущего треда, то можно поступить так:
df = pd.Series(data['cast'].str.cat(sep='|').split('|')).value_counts()
for i in df.index:
    df[i] = data['revenue'][data['release_year']==2012][data['cast'].map(lambda x: True if i in x else False)].sum()
df = pd.DataFrame(df).sort_values(0, ascending=False)
df
```

Нужно сначала разделить имена актеров по делителю '|', вы можете воспользоваться алгоритмом из треда выше.
И уже потом для каждого из них считать revenue.sum() в отдельности

# Вопрос 16. Какой актер снялся в большем количестве высокобюджетных фильмов?

```python
def counter(movie_bd, x):
    data_plot = movie_bd[x].str.cat(sep='|')
    dat = pd.Series(data_plot.split('|'))
    info = dat.value_counts(ascending=False)
    return info
sum_gen = counter(data[data['budget'] > data['budget'].mean()], 'cast')
sum_gen
```
